```python
import bs4
from langchain import hub
from langchain_chroma import Chroma
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAIEmbeddings
from langchain_community.document_loaders import WebBaseLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", api_key="...")
```

```python
loader = WebBaseLoader(
    web_paths=("https://spartacodingclub.kr/blog/all-in-challenge_winner",),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("editedContent")
        )
    ),
)
docs = loader.load()
```

```python
docs
```

```python
# 수상작 하나의 내용이 잘릴수 있으니 chunk_size 증가시킴
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=10000,
    chunk_overlap=1000
)
splits = text_splitter.split_documents(docs)

vectorstore = Chroma.from_documents(
    documents=splits,
    embedding=OpenAIEmbeddings(api_key="...")
)
```

```python
from langchain_core.prompts import PromptTemplate
#  retriever의 검색 문서 수 늘리기 (k 조절, 기본 k=4)
retriever = vectorstore.as_retriever(search_kwargs=dict(k=20))

user_msg = "ALL-in 코딩 공모전 수상작 전체 내용에서 수상작명과 서비스 기능을 요약해줘"

# retriever.invoke()는 LLM에게 전달할 문서를 찾는 단계
# PromptTemplate이 출력 형식이나 요약 방식을 지정하는 핵심
# 그래서 user_msg를 아무리 예쁘게 포맷해도, LLM이 어떻게 응답할지는 Prompt에서 결정
retrieved_docs = retriever.invoke(user_msg)

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
# prompt = hub.pull("rlm/rag-prompt")

template = """
당신은 정보 요약 도우미입니다.
다음의 문서들을 참고하여, 각 수상작에 대해 다음 형식으로 정리해 주세요:

수상작명:
서비스 기능:

아래는 문서 내용입니다:
--------------------------
{context}
--------------------------

결과는 수상작별로 나열해 주세요.
반드시 형식을 지켜 주세요.
"""

prompt = PromptTemplate.from_template(template)

user_prompt = prompt.invoke({"context": format_docs(retrieved_docs)})
print(user_prompt)
```

```python
response = llm.invoke(user_prompt)
print(response.content)
```
